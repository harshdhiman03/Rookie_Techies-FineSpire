```python
from datasets import load_dataset

dataset = load_dataset("gbharti/finance-alpaca",split='train')
```

```python
dataset.to_pandas()
```

```python
print(dataset.features)
```

```python
def format_data(example):
    input_text = f"Task: Answer the following financial question.\nQuestion: {example['input']}"
    example['instruction'] = input_text
    example['output'] = example['input']
    return example

formatted_dataset = dataset.map(format_data)
```

```python
formatted_dataset = formatted_dataset.train_test_split(test_size=.2)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('t5-small')
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples['text'],
        padding="max_length",
        truncation=True,
        max_length=128
    )

tokenized_datasets = formatted_dataset.map(tokenize_function, batched=True)
```

```python
label_arrays = [example['instruction'] for example in tokenized_datasets['train']]
label_arrayst = [example['output'] for example in tokenized_datasets['test']]

print(label_arrays[:5],label_arrayst[:5])
```

```python
prefix = 'answer: '

def process_function(examples):
  # Check if 'text' is the correct column name for your documents
  inputs = [prefix + doc for doc in examples['text']]
  model_inputs = tokenizer(inputs, max_length=1024, truncation=True, padding='max_length')

  with tokenizer.as_target_tokenizer():
      # Assuming 'output' is the correct column name for your summaries
      labels = tokenizer(examples['output'], max_length=128, truncation=True, padding='max_length')

  model_inputs['labels'] = labels['input_ids']
  return model_inputs
```

```python
tokenized_dataset = tokenized_datasets.map(process_function, batched = True)
```

```python
from transformers import DataCollatorForSeq2Seq, AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
data_collator = DataCollatorForSeq2Seq(tokenizer= tokenizer, model= 't5-small')
model = AutoModelForSeq2SeqLM.from_pretrained('t5-small')
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir = './results',
    evaluation_strategy = 'epoch',
    learning_rate = 2e-5,
    per_device_train_batch_size = 10,
    per_device_eval_batch_size = 10,
    weight_decay = 0.01,
    save_total_limit = 3,
    num_train_epochs = 2,
    # predict_with_generate = True,
    fp16 = True,
    # push_to_hub = True,
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],
    eval_dataset=tokenized_dataset['test'],
    data_collator=data_collator,
)
```

```python
# prompt: make a flask api for this model

!pip install flask

# Create a Flask app
from flask import Flask, request, jsonify
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

app = Flask(__name__)

# Load your trained model and tokenizer
# Replace 'trained_model_state.pth' with the actual path to your saved model state
model_state = torch.load('trained_model_state.pth')
model = AutoModelForSeq2SeqLM.from_pretrained('t5-small')  # Load the base model
model.load_state_dict(model_state)
model.eval()  # Set the model to evaluation mode

tokenizer = AutoTokenizer.from_pretrained('t5-small')

# Define a route for prediction
@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    text = data.get('text')

    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt")

    # Generate prediction
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128)

    # Decode the generated output
    prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return jsonify({'prediction': prediction})

if __name__ == '__main__':
    app.run(debug=True)
```

```python
!pip install chainlit langchain transformers yfinance

import chainlit
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline
import yfinance as yf

# Load the model from Hugging Face Hub (or locally)
model_name = "your-username/your-model-name"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Create a pipeline for text generation
pipe = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=128
)

# Wrap the pipeline in a LangChain LLM
llm = HuggingFacePipeline(pipeline=pipe)

# Define a prompt template that includes stock information
prompt_template = """Task: Answer the following financial question.
Stock Information: {stock_info}
Question: {question}"""
prompt = PromptTemplate(template=prompt_template, input_variables=["stock_info", "question"])

# Create an LLM chain
chain = LLMChain(llm=llm, prompt=prompt)

# Define Chainlit integration
@on_message
async def main(message: str):
    # Fetch stock data
    ticker = "AAPL"  # Replace with the desired stock ticker
    stock_data = yf.Ticker(ticker)
    current_price = stock_data.info['currentPrice']

    # Format stock information for the prompt
    stock_info = f"The current price of {ticker} is {current_price}."

    # Run the LLM chain with the stock information
    response = chain.run(stock_info=stock_info, question=message)
    await Chainlit.Message(content=response).send()

# Run Chainlit
Chainlit.run()
```
